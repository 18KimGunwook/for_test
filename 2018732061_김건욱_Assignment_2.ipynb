## Problem 1
* 강의자료 예제 실행해보기  
  + 강의자료의 **Python 기초 문법** 예제들을 이 아래에 코드블럭으로 재작성하고 실행



```python
#한 줄 주석처리는 다음과 같이 #"문장" 으로 처리한다.
# helloworld
print('hello world')
```

```python
#작은 따음표 3개(''') 로 문장의 앞뒤를 둘러쌓아 주석 블럭으로 처리한다.
'''
Function name: helloworld()
This function prints 'hello world'
'''
def helloworld():
    print('hello world')

helloworld()
```

```python
#다음은 0~9 까지의 숫자를 5와 비교하는 코드 이다.
#코드의 시작을 알린다.
print('start')
#변수 i 를 0부터 시작하여 9까지 반복하게 한다.
for i in range(10):
    #5 초과이면 i > 5를 출력하게 한다.
    if i>5:
        print('i = %d > 5' % i)
    #5 미만이면 i < 5를 출력하게 한다.
    elif i<5:
        print('i = %d < 5' % i)
    #5 와 동일하면 i = 5를 출력하게 한다.
    else:
        print('i = %d = 5' % i)
#코드의 끝을 알린다.
print('end')
```

```python
#함수의 사용예시를 나타내는 코드이다.
#함수의 정의 선언부이다.
def function1():
    #함수가 시작하고 끝남을 출력하게 한다.
    print('  function start')
    print('  function end')

#코드 실행시 스크립트 시작과 끝 사이에
#함수의 시작과 끝이 들어가짐을 확인 가능하다
print('script start')
function1()
print('script end')
```

```python
#함수의 입력 예시이다.
#변수 a에 문자열이 입력되고 이를 출력한다.
a = input()
print(a)
```

```python
#각기 다른 변수형을 출력하는 코드 이다.
#i는 정수이므로 %d로 출력한다.
i = 4
#j는 실수이므로 %f로 출력한다.
j = 4.5
#k는 문자열이므로 %s로 출력한다.
k = 'abc'
#최종적으로 변수들을 출력한다.
print('i=%d j=%.1f k=%s' % (i,j,k))
```

## Problem 2
* 프로그램 작성해보기
  + 제시된 출력결과를 보고, 아래 미완성 프로그램을 완성



```python
#name을 입력받는다.
print("What is your name?")
name = input()
#age를 입력받는다.
print("How old are you?")
age = input()
#grade를 입력받는다.
print("What is your grade point average?")
grade = input()

#name, age, grade를 각각 출력한다.
#name, age, grade는 별다른 코드 없이 input으로 입력받는다면
#문자열로 저장되기 때문에 %s로 출력한다.
#만일 다른 변수형으로 입력받고 싶다면
#int(input())과 같이 변수형으로 input()을 둘러싸게 한다.
print('Your name is %s' % name)
print('Your age is %s' % age)
print('Your grade is %s' % grade)
```
