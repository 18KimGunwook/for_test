## Logistic Regression

* Data 파일을 이용한 Logistic Regression
  + data_logistic_regression.csv을 이용해 학습한다
    + 8개의 input, 1개의 output  
$$ [x_1,...x_8,y] $$
* Accuracy 75% 이상을 달성한다

```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as func
import torch.optim as opt
import csv

# Seed 고정
torch.manual_seed(1)
```

```python
# csv 파일로 부터 데이터를 불러온다
with open('data_logistic_regression.csv') as file_name:
  # array 변수에 shape가 (759,9)인 데이터를 입력한다.
    array = np.loadtxt(file_name, delimiter=",")

#x_train은 1~8열 까지의 데이터를 입력한다.
x_train = torch.FloatTensor(array[:,0:8])
#y_train은 9열의 데이터를 입력한다.
y_train = torch.FloatTensor(array[:,8:9])
```

```python
# 모델생성전 적절한 배열로 입력이 되었는지를 확인한다.
print(x_train.shape)
print(y_train.shape)

# sigmoid 함수로 Model을 설계한다.
class Logistic(nn.Module):
    def __init__(self, num_inputs, num_outputs):
        super().__init__()
        self.linear = nn.Linear(num_inputs, num_outputs)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        return self.sigmoid(self.linear(x))

# Model 초기화 (입력 8, 출력 1)
model = Logistic(8, 1)

```

```python
# Optimizer 설정 (learning rate = 1로 설정)
optimizer = opt.SGD(model.parameters(), lr=1)

# 100번 반복한다.
for epoch in range(100):

  # Cost 계산 / mse_loss(가정에의한값, 참값)
  y_hypo = model(x_train)
  cost = func.binary_cross_entropy(y_hypo, y_train)

  # cost를 이용해 model update
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  # 10번 마다 중간결과 출력
  if epoch % 10 == 9:
    prediction = y_hypo >= torch.FloatTensor([0.5])
    correct_prediction = prediction.float() == y_train
    accuracy = correct_prediction.sum().item() / len(correct_prediction)
    print('Epoch {:4d}/{} Cost: {:.6f} Accuracy {:2.2f}%'.format(
            epoch, 100, cost.item(), accuracy * 100,
    ))
```
