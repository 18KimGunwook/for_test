## Multinomial Classification

* Data 파일을 이용한 Multinomial Classification
  * data_multinomial_classification.csv을 이용해 학습한다
    * 16개의 input, 1개의 output
* Accuracy 95% 이상을 달성한다

```python
#Import

import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as func
import torch.optim as opt
import csv

# Seed 고정
torch.manual_seed(1)
```

```python
# csv 파일로 부터 데이터를 불러온다
with open('data_multinomial_classification.csv') as file_name:
  # array 변수에 shape가 (759,9)인 데이터를 입력한다.
    array = np.loadtxt(file_name, delimiter=",")

#x_train은 1~16열 까지의 데이터를 입력한다.
x_train = torch.FloatTensor(array[:,0:16])
#y_train은 0~6의 정수를 가지는 17열의 데이터를 입력한다.
y_train = torch.LongTensor(array[:,16])
```

```python
print(x_train.shape)
print(y_train.shape)

# Model 초기화 (입력 dim, 출력 dim)
model = nn.Linear(16,7)
```

```python
# Optimizer 설정 (learning rate = 0.1로 설정)
optimizer = opt.SGD(model.parameters(), lr=0.1)

# 반복
for epoch in range(1000):

  # Cost 계산 / mse_loss(가정에의한값, 참값)
  y_hypo = model(x_train)
  cost = func.cross_entropy(y_hypo, y_train)

  # cost를 이용해 model update
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  # 10번 마다 중간결과 출력
  if epoch % 100 == 99:
    correct_prediction = y_hypo.argmax(dim=1) == y_train
    accuracy = correct_prediction.sum().item() / len(correct_prediction)
    print('Epoch {:4d}/{} Cost: {:.6f} Accuracy {:2.2f}%'.format(
            epoch, 1000, cost.item(), accuracy * 100,
    ))
```
