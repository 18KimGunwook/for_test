#Numpy 실습

* 행렬방정식 풀기
  + 다음 행렬 방정식을 'Pseudo inverse matrix'를 이용해 풀어보자
  + $A^TA$의 역행렬이 존재한다고 가정

  $Ax=B$

  $A=\begin{bmatrix}0&1\\1&1\\2&1\\3&1\\ \end{bmatrix}$  
  $B=\begin{bmatrix}-1\\0.2\\0.9\\2.1\\ \end{bmatrix}$

---
* Pseudo Inverse matrix  

  $A^+=(A^TA)^{-1}A^T$  
  $x=A^+B$

```python
import numpy as np
import numpy.linalg as lin

#행렬 A를 선언 및 출력한다.
A_mat=np.array([[0, 1],[1, 1],[2, 1],[3, 1]])
print('A=')
print(A_mat)

#행렬 B를 선언 및 출력한다.
B_mat=np.array([[-1],[0.2],[0.9],[2.1]])
print('B='),
print(B_mat)

#A의 유사역행렬을 위의 공식을 통해 계산한다.
pseudo_inv_A_mat=np.dot(lin.inv(np.dot(A_mat.T,A_mat)),A_mat.T)
print('Pseudo Inverse A=')
print(pseudo_inv_A_mat)

#유사역행렬을 통해 행렬 x를 구한다.
x_mat=np.dot(pseudo_inv_A_mat,B_mat)
print('x='),
print(x_mat)
```

# Matplotlib 실습  

* 2x2 그래프 그리기    
  + $y=x$  
  + $y=x^2$  
  + $y=sin(x)$
  + $y=cos(x)$


```python
import numpy as np
import matplotlib.pyplot as plt

#x축의 범위를 설정한다.
x = np.arange(0, 3 * np.pi, 0.1)

#출력할 함수들을 설정한다.
y_x=x
y_pow=pow(x,2)
y_sin=y=np.sin(x)
y_cos=y=np.cos(x)

plt.clf()
# 그래프 간격 조절한다.
left=0.0
bottom=0.0
right=1.5
top=1.0
wspace=0.3
hspace=1.0

plt.subplots_adjust(left, bottom, right, top, wspace, hspace)

#2*2 그래프이므로 plt.subplot(2,2,n)으로 좌표를 입력한다.
plt.subplot(2,2,1)
plt.plot(x,y_x)
plt.title(r'$y = x$')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.subplot(2,2,2)
plt.plot(x,y_pow)
plt.title(r'$y=x^2$')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()

#밑에서 한번더 그래프 간격을 조절해줘야 정상적으로 그래프가 출력된다.
plt.clf()
plt.subplots_adjust(left, bottom, right, top, wspace, hspace)

plt.subplot(2,2,3)
plt.plot(x,y_sin)
plt.title(r'$y=sin(x)$')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.subplot(2,2,4)
plt.plot(x,y_cos)
plt.title(r'$y=cos(x)$')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
