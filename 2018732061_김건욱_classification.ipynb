```python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
import os
import numpy as np

files = [[0 for j in range(5)] for i in range(120)]

for i in range(4):
  path = '/content/drive/MyDrive/'+str(i+1) # 폴더 경로
  os.chdir(path) # 해당 폴더로 이동
  files[i] = os.listdir(path)
```

```python
from tensorflow.python.ops.math_ops import sobol_sample

from google.colab.patches import cv2_imshow
import numpy as np
import keras.utils as image
import matplotlib.pyplot as plt
import cv2

X_train=[]

for i in range(5):
  for j in range(120): 
        f = cv2.imread('/content/drive/MyDrive/'+str(i+1)+'/'+str(files[i][j]),cv2.IMREAD_COLOR)
        f_resize= cv2.resize(f, (256,256), interpolation=cv2.INTER_LINEAR)
        X_train.append(f_resize)

```

```python
category_var=np.array([0,1,2,3,4])
b = np.zeros((category_var.size, 5)) 
b[np.arange(category_var.size),category_var] = 1

Y_train=np.zeros((600,5))
Y_test=np.zeros((100,5))

for i in range(5):
  Y_train[i*120:i*120+120,:]=b[i]
  Y_test[i*20:i*20+20,:]=b[i]

```

```python
import random

X_test=[]

for i in range (5):
  for j in range(20):
    f=f = cv2.imread('/content/drive/MyDrive/'+str(i+1)+'/'+str(files[i][random.randrange(0,120)]),cv2.IMREAD_COLOR)
    f_resize= cv2.resize(f, (256,256), interpolation=cv2.INTER_LINEAR)
    X_test.append(f_resize)
```

```python
X_train = X_train/255.0
X_test = X_test/255.0

Y_train = to_categorical(Y_train)
Y_test = to_categorical(Y_test)
```

```python
model = Sequential()
model.add(Conv2D(64, kernel_size=(3, 3), input_shape=(32, 32, 3), activation='relu',padding='same'))
model.add(MaxPooling2D(pool_size=2,padding='same'))
model.add(Conv2D(128, (3, 3), activation='relu',padding='same'))
model.add(MaxPooling2D(pool_size=2,padding='same'))
model.add(Conv2D(256, (3, 3), activation='relu',padding='same'))
model.add(MaxPooling2D(pool_size=2,padding='same'))
model.add(Flatten())
model.add(Dense(5, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=2E-04),
              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, Y_train, epochs=50, batch_size=200)

# 테스트 정확도 출력
print("\n Test Accuracy: %.4f" % (model.evaluate(X_test, Y_test)[1]))
```
