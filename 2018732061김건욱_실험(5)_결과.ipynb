```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as func
import torch.optim as opt

# Seed 고정
torch.manual_seed(1)
```

```python
#input은 x_train.shape = (8,4) 이므로 4이다.
x_train = torch.FloatTensor([[1, 2, 1, 1], [2, 1, 3, 2], [3, 1, 3, 4], [4, 1, 5, 5],
               [1, 7, 5, 5], [1, 2, 5, 6], [1, 6, 6, 6], [1, 7, 7, 7]])
#output은 0~2의 숫자이므로 3이다.
y_train = torch.LongTensor([2, 2, 2, 1, 1, 1, 0, 0])

print(x_train)
print(x_train.shape)
print(y_train)
print(y_train.shape)

#cross entropy에서 sigmoid 함수를 만들어주므로 Linear 모델만 만들어 준다.
#input, output은 위에서 설명한대로 4,3이다.
model = nn.Linear(4, 3)
```

```python
#learning rate = 0.1 
optimizer = opt.SGD(model.parameters(), lr=0.1)

#gradient descent를 1000번 실행
for epoch in range(1000):

  #y_hypo를 설정한다.
  y_hypo = model(x_train)
  #가중치를 cross entropy로 설정 및 sigmoid 함수를 세운다.
  cost = func.cross_entropy(y_hypo, y_train)

  #gradient descent 실행한다.
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  #100번 마다 출력한다.
  if epoch % 100 == 99:
    #correct_prediction를 y_hypo에 hot_encoding을 실행한 값으로 설정함과 동시에
    #y_train과 결과값을 비교한다.
    correct_prediction = y_hypo.argmax(dim=1) == y_train
    #정확도를 계산한다.
    accuracy = correct_prediction.sum().item() / len(correct_prediction)
    #가중치와 정확도를 출력한다.
    print('Epoch {:4d}/{} Cost: {:.6f} Accuracy {:2.2f}%'.format(
            epoch, 1000, cost.item(), accuracy * 100,
    ))
```

```python
dataset = np.loadtxt(
    '/content/drive/MyDrive/Colab Notebooks/data_multinomial_classification.csv', 
    delimiter=',', 
    dtype=np.float32)

#x0~x15 까지 존재하므로 input은 16이다.
x_train = torch.FloatTensor(dataset[:,:-1])
#y_train은 0~6까지의 정수들 이므로 output은 7이다.
y_train = torch.LongTensor(dataset[:,-1])
model = nn.Linear(16, 7)

#learning rate = 0.1
optimizer = opt.SGD(model.parameters(), lr=0.1)

#이하는 위의 실습과 동일하다.
for epoch in range(1000):

  y_hypo = model(x_train)
  cost = func.cross_entropy(y_hypo, y_train)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch % 100 == 99:
    correct_prediction = y_hypo.argmax(dim=1) == y_train
    accuracy = correct_prediction.sum().item() / len(correct_prediction)
    #정확도가 100%에 도달함을 알 수 있다.
    print('Epoch {:4d}/{} Cost: {:.6f} Accuracy {:2.2f}%'.format(
            epoch, 1000, cost.item(), accuracy * 100,
    ))
```
