###3.이론  

####Multinomial Classification  
***

실험 4 에서 Logistic Regression를 통해 데이터를 0,1 로만 결과값으로 나타내었다. 하지만 Multinomial Classification는 더 다양한 값들로 결과값을 분류한다.  
따라서 이를 sigmoid로 나타내면 밑과 같은 결과가 나온다.  

![image.png](https://i.ibb.co/sQxhjDJ/multinomial-sigmoid.png)  
그러나 이경우에 양 끝단의 값과 그 사이의 값의 x축 구간이 차이나게 되는 문제가 발생하게 된다. 따라서 밑에서 설명할 Softmax를 사용하게 된다.   

####Softmax  
***

Softmax란 n개의 클래스가 있을 때, n차원의 벡터를 입력받아 결과값을 각 클래스의 속할 **확률**로 나타내는 것이다. 이 성질때문에 결과값의 총합은 1이 된다.  
Softmax의 함수 공식은 다음과 같다.

$$
y_k = \frac{e^{a_k}}{\sum_{i=1}^{n}e^{a_i}}
$$

자연상수 e를 사용하여 확률을 나타냈기 때문에 결과값이 부정확 할 수 있으나, 미분이 가능하다는 큰 장점을 가지게 된다.  

####One-Hot Encoding 
***
위의 Softmax를 통해 Hypothesis를 한다면 이에 대한 cost가 필효한데 이 때 쓰이는 함수가 One-Hot Encoding 이다.  
One-Hot Encoding 이란 클래스에서 가장 큰 값을 1로 만들고 나머지를 0으로 만드는 함수이다. 이를 확률적으로 해석하면 하나의 값이 나올 확률이 100% 이며 나머지 경우는 0%임을 뜻하기 때문에, 가장 **이상**적인 경우라 볼 수 있고 따라서 이를 cost로 활용하는 것이다.  

####Markdown에서 구현  
***

Multinomial Classification의 실행을 위해 y_hypo 의 값을 Softmax(Linear(x))로 두어야 하나, `func.cross_entropy()` 함수에서 softmax가 포함되어 있으므로 model은 linear model로 사용한다.  
prediction은 y_hypo 에 `argmax(dim=1)` 함수를 취해 One-Hot Encoding로 나타낸 값을 y_train 값과 비교해 accuracy를 구한다.
