```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.keras as K

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

data=np.loadtxt("1.txt",delimiter=",")

X=data[:,0]
Y=data[:,1]

X_new=np.column_stack((X*X,X))

linear_model=K.Sequential()
linear_model.add(K.layers.Dense(1,input_dim=2))
linear_model.compile(optimizer=K.optimizers.SGD(learning_rate=1E-03), loss='mean_squared_error', metrics=['mse'])

history=linear_model.fit(X_new,Y,epochs=1000)

fig=plt.figure()
plt.scatter(X,Y,c='g',label='Real data')
plt.plot(X,linear_model(X_new),c='r',label='hypothesis')
plt.legend(loc='upper right')
```
