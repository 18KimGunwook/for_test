```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.keras as K

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical

import pandas as pd

data=pd.read_excel('baseball.xlsx',dtype='float32')
data=data.values

X=data[:,:5]
Y=data[:,5]
Y = to_categorical(Y)

model= Sequential()
model.add(Dense(3, activation='softmax', input_dim=5))
model.compile(optimizer=Adam(learning_rate=3), loss='categorical_crossentropy', metrics=['accuracy'])

history= model.fit(X,Y,epochs=3000)

X_test = X
Y_test = model.predict(X_test)

print(model.metrics_names)
print(model.evaluate(X_test, Y_test))

plt.plot(history.history['loss'])
plt.title('Model error')
plt.xlabel('Epoch')
plt.legend('Loss',loc='upper right')
```
